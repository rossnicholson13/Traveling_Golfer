```python
import pandas as pd
```

```python
final_db = pd.read_csv('final_db.csv')
```

```python
final_db.head()
```

```python
final_db = final_db.rename(columns={'Tee Name': 'Tee', 'Slope Rating (18)': 'Slope_Rating', 'Bogey Rating (18)': 'Bogey_Rating', 'USGA_Rating': 'Usgarating'})
```

```python
final_db.to_csv('final_db.csv')
```

```python
final_db.head(35)
```

```python
names = []
lat = []
lng = []
info = []

for i in range(len(final_db)):
    if i == 0:
        names.append(final_db.iloc[i,3])
        lat.append(final_db.iloc[i,8])
        lng.append(final_db.iloc[i,9])
        
    if i <= len(final_db) - 2 and i > 0: 
        if final_db.iloc[i, 6] != final_db.iloc[i-1, 6]:
            names.append(final_db.iloc[i,3])
            lat.append(final_db.iloc[i,8])
            lng.append(final_db.iloc[i,9]) 
            
        if (final_db.iloc[i, 6] != final_db.iloc[i+1, 6]):
            temp_df = final_db.loc[final_db['Course ID'] == final_db.iloc[i,6]]
            course_info = {
                'Tee': list(temp_df['Tee']),
                'Par': list(temp_df['Par']),
                'USGA_Rating': list(temp_df['Usgarating']),
                'Slope_Rating': list(temp_df['Slope_Rating']),
                'Bogey_Rating': list(temp_df['Bogey_Rating']),
                'Gender': list(temp_df['Gender'])
            }
            info.append(course_info)
        
    if i == len(final_db) - 1:
        temp_df = final_db.loc[final_db['Course ID'] == final_db.iloc[i,6]]
        course_info = {
            'Tee': list(temp_df['Tee']),
            'Par': list(temp_df['Par']),
            'USGA_Rating': list(temp_df['Usgarating']),
            'Slope_Rating': list(temp_df['Slope_Rating']),
            'Bogey_Rating': list(temp_df['Bogey_Rating']),
            'Gender': list(temp_df['Gender'])
        }
        info.append(course_info)
    
```

```python
course_db = pd.DataFrame({
    "Name": names,
    "Latitude": lat,
    "Longitude": lng,
    "Info": info
})
```
